```python
#Catégorisation de la durée de crédit en se basant sur les ecarts inter quartiles
def categorize_Duration(duration):
    if duration<12:
        return 1
    elif 12 <= duration < 18:
        return 2
    elif 18 <= duration < 24:
        return 3
    else:
        return 4
        
#data['Duration_Category']=data['Duration_of_Credit_monthly'].apply(categorize_Duration)
#data[['Duration_of_Credit_monthly','Duration_Category']].head(20)
```

```python
#Catégorisation de l'age en se basant sur les écarts inter quartiles
def categorize_age (age):
    if age <27:
        return 1
    elif age>=27 and age<33 :
        return 2
    elif age>=33 and age<37:
        return 3
    else:
        return 4
    

#data['Age_category']=data['Age_years'].apply(categorize_age)
#data[['Age_years','Age_category']].head(10)
```

```python
#Categorisation du montant de crédit suivant les écarts inter quartiles
def categorize_amount(montant):
    if montant <1365:
        return 1
    elif montant >=1365 and montant<2319:
        return 2
    elif montant >=2319 and montant<3972:
        return 3
    else:
        return 4
    

#data['Mount_Category']=data['Credit_Amount'].apply(categorize_amount)
#data[['Mount_Category','Credit_Amount']].head(10)
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('duration_cat', FunctionTransformer(lambda x: x.applymap(categorize_Duration)), ['Duration_of_Credit_monthly']),
        ('age_cat', FunctionTransformer(lambda x: x.applymap(categorize_age)), ['Age_years']),
        ('amount_cat', FunctionTransformer(lambda x: x.applymap(categorize_amount)), ['Credit_Amount']),
    ],
    remainder='passthrough'  # Keeps the other columns unchanged
)

# Create the pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])

# Assuming `data` is your DataFrame
# Apply the pipeline to your data
processed_data = pipeline.fit_transform(data)

# Create a new DataFrame with the transformed columns
transformed_columns = ['Duration_Category', 'Age_category', 'Mount_Category']
original_columns = list(data.columns.drop(['Duration_of_Credit_monthly', 'Age_years', 'Credit_Amount']))
all_columns = transformed_columns + original_columns

processed_df = pd.DataFrame(processed_data, columns=all_columns)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class CramersVTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, target_col):
        self.target_col = target_col
    
    def cramers_v(self, confusion_matrix):
        chi2 = chi2_contingency(confusion_matrix)[0]
        n = confusion_matrix.sum()
        phi2 = chi2 / n
        r, k = confusion_matrix.shape
        phi2corr = max(0, phi2 - ((k-1)*(r-1)) / (n-1))
        rcorr = r - ((r-1)**2) / (n-1)
        kcorr = k - ((k-1)**2) / (n-1)
        cramers_v_value = np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
        return cramers_v_value

    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        categorical_vars = X.select_dtypes(include='category').columns
        crammers_v_values = []
        for var in categorical_vars:
            if var != self.target_col:
                confusion_matrix = pd.crosstab(data[self.target_col], X[var])
                v = self.cramers_v(confusion_matrix.to_numpy())
                crammers_v_values.append((var, v))
        crammers_v_values.sort(key=lambda x: x[1], reverse=True)
        cramers_v_df = pd.DataFrame(crammers_v_values, columns=['Variable', 'V de Cramer'])
        
        # Visualization
        plt.figure(figsize=(10, 8))
        plt.barh(cramers_v_df['Variable'], cramers_v_df['V de Cramer'], color='skyblue')
        plt.xlabel('V de Cramer')
        plt.ylabel('Variable')
        plt.title('Pouvoir Discriminant des Variables (V de Cramer)')
        plt.gca().invert_yaxis()  # Inversement l'ordre des variables pour que la plus discriminante soit en haut
        plt.show()
        
        return X

# Define the full pipeline
full_pipeline = Pipeline(steps=[
    ('cramers_v', CramersVTransformer(target_col='Creditability'))
])

# Apply the pipeline to your data
processed_data = full_pipeline.fit_transform(processed_df)

# Output the processed data for inspection
processed_data
```
